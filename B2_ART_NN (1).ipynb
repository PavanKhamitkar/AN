An ART network has 2 nodes in the input layer, interface layer and the output layer. The bottom-up weights are given by u = [03,05;0.2,0.6] and top-down weights are d = [0.1,0.3;0.6,0.1]. Take alpha =4 and the vigilance parameter, v= 0.3. The network is given an input S = [0.5,0.6]. Find the cluster to which S belongsART_diagram.PNG

```python
import numpy as np
```

**Feed the input parameters** to the network

```python
S = [0.5,0.6]
u = [[0.3,0.5],[0.2,0.6]]
d = [[0.1,0.3],[0.6,0.1]]
v = 0.3
print('The input pattern is=',S)
print('The bottom up weights are =',u)
print('The top-down weights are =',d)
```

Here, the L1 norm is calculated by summing up the individual elements of S,input vector. L1 norm is also called as Manhattan distance

**STEP 1:** **Check the** **winning cluster** by calculating O1 and O2

```python
# The input layer's norm calculation
Sn = np.linalg.norm(S[0]+S[1])
print(Sn)
I = S
O1 = I[0]*u[0][0] + I[1]*u[0][1]
print(O1)
O2 = I[0]*u[1][0] + I[1]*u[1][1]
print(O2)
```

**STEP 2:** Decide the winning cluster by comparing O1 AND O2

```python
if (O1<O2):
  print('O1 is the winning cluster')
  I1 = S[0]*d[0][0]
  I2 = S[1]*d[0][1]
  print(I1)
  print(I2)
else:
  print('O2 is the winning cluster')
  I1 = S[0]*d[1][0]
  I2 = S[1]*d[1][1]
  print(I1)
  print(I2)
```

**STEP 3:** Calculate the **NORM OR Matching parameter** of input nodes/ neurons

```python
In = np.linalg.norm(I1+I2)
print(In)
Match_factor = In/Sn
print('Match parameter ', Match_factor)
```

**STEP 4:** Compare the MATCH factor with **Vigilance parameter** and update the I1 and I2 norms wrt to winning cluster

```python
if Match_factor<v:
   print('Cluster 2 is selected')
   I1 = S[0]*d[1][0]
   I2 = S[1]*d[1][1]
   In1 = np.linalg.norm(I1+I2)
   MF = In1/Sn
   print(MF)
```

**STEP 5:** Last step of verification in the recognition stage

```python
if MF > v:
  print('The input pattern "S" is assigned to cluster 2')
else:
  print('The input pattern does not belong to any of the clusters:Choose to change the vigilance parameter')
```
